```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None) 
```

```python
Pop = pd.read_csv("Population.csv", delimiter=",")
ManageDrinking = pd.read_csv("BasicAndSafelyManagedDrinkingWaterServices.csv", delimiter=",")
Mortality = pd.read_csv("MortalityRate copie.csv", delimiter=";")
Politic =  pd.read_csv("PoliticalStability.csv", delimiter=",")
Region = pd.read_csv("RegionCountry.csv", delimiter=",")
#UrbanRate = pd.read_csv("Complete2.csv", delimiter=",")
#display(Region[:10], Region.shape)
```

### Nettoyage fichier Country Regions

```python
Region.rename(columns={"COUNTRY (DISPLAY)": "Country", "REGION (DISPLAY)": "Region"}, inplace=True)
print("Shape fichier region",Region.shape)
print("Valeurs uniques du fichier region",'\n',Region.nunique())
```

Le fichier Region possede 194 lignes pour 194 Country et 6 region. Chaque pays n'apparait donc qu'une seule fois

### Fichier Population

```python
# Supression des lignes "Male" et "Female"

Pop = Pop.loc[Pop["Granularity"]!= "Male"]
Pop = Pop.loc[Pop["Granularity"]!= "Female"]

print("Shape de pop:",Pop.shape)
print(Pop.nunique())
```

### Fusion Fichier population et regions

```python
Complete_Df = pd.merge(Pop, Region, on="Country", how="right")
display(Complete_Df[:1])
print("Shape de complete",Complete_Df.shape,'\n')
print("Valeurs uniques de Complete_df",'\n',Complete_Df.nunique())
```

### Fichier Politic stability

```python

Politic.info()
Politic.nunique()

```

le fichhier de stabilité politique traite 18 années (entre 2000 et 2018). Atention il manque l'année 2001.

### Fusion de Complete_DF Political Stability

```python
Complete_Df= pd.merge(Complete_Df, Politic, on=["Country", "Year", "Granularity"], how="left")

#Complete_Df.rename(columns={"Political_Stability_x": "Political_Stability"}, inplace=True)
#Complete_Df.drop(labels="Political_Stability_y", axis=1,inplace=True)
display(Complete_Df[:5])
print("Shape de Complet_df",Complete_Df.shape)
print(Complete_Df.nunique())
```

Apres le Merge les pays en plus qui etait present dans le fichier de stabilite politique ont été supprimé

### Fichier Mortality Rate

```python
print("Shape Mortality Rate",Mortality.shape)
print("\nValeur unique dans Mortality rate",'\n',Mortality.nunique())
display(Mortality[:1])
print(Complete_Df.nunique())
```

### Fusion entre Complete_Df et le fichier "Mortality rate"

```python
Complete_Df= pd.merge(Complete_Df, Mortality, on=["Country", "Year", "Granularity"], how="left")
#Complete_Df.drop(labels="Political_Stability_y", axis=1,inplace=True)
#display(Complete_Df.loc[Complete_Df["Year"]== 2016])
print(Complete_Df.shape)
display(Complete_Df[:1])
print(Complete_Df.nunique())
print(Complete_Df.isna())

```

### Fichier Basic and safely managed watter drinking services

```python
print("Shape ManageDrinking",ManageDrinking.shape)
print("\nValeur unique dans ManegeDrinking",'\n',ManageDrinking.nunique())
display(ManageDrinking[:3])
```

### Fusion entre Complete et Managed water drinking services

```python
print(Complete_Df.nunique())
Complete_Df= pd.merge(Complete_Df, ManageDrinking, on=["Country", "Year", "Granularity"], how="left")
#Complete_Df.drop(labels="Political_Stability_y", axis=1,inplace=True)

print(Complete_Df.shape)
display(Complete_Df[:3])
print(Complete_Df.nunique())
```

### Calcul du taux d'habitant en zone urbaine par pays et par années

```python
# Recuperation du nombre d'habitant zone urbaine par pays et par années
urban_rate= Complete_Df.loc[ Complete_Df["Granularity"] == "Urban"] 

# Recuperation du nombre d'habitant total par pays et par années
Pop.drop_duplicates(subset=["Country","Year"], inplace=True)
display(Pop[:5])

#Creation d'un DataFrame ou reunir toutes les valeurs
urban_rate2 = pd.merge(urban_rate, Pop, on=["Country","Year"], how="left")

#Calcul du taux d'habitant en zone urbaine 
urban_rate2["Rate pop urban"] = round((urban_rate2["Population_x"] * 100) / urban_rate2["Population_y"], 1)


display(urban_rate2[:1])
urban_rate2.shape
```

```python
# Nettoyage de urban rate pour n'avoir plus que les colonne a fusionner avec complete_df

urban_rate2.drop(['Political_Stability', 'Region',"Mortality rate attributed to exposure to unsafe WASH services",
"WASH d","Population using at least basic drinking-water services (%)","Granularity_y",
"Population using safely managed drinking-water services (%)","Population_x","Population_y"], axis=1, inplace=True)
urban_rate2.rename(columns={"Granularity_x":"Granularity"},inplace=True)
display(urban_rate2[:5])
print(Complete_Df.nunique())
```

### ajout de la colonne du taux de la population vivant en zone urbaine("rate pop urban") a completeDf

```python
Complete_Df= pd.merge(Complete_Df, urban_rate2, on=["Country", "Year", "Granularity"], how="left")
display(Complete_Df[:10])
print(Complete_Df.shape)
```
